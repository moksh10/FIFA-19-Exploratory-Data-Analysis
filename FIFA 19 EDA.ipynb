# FIFA Exploratory Data Analysis

```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

## TASK 1: READ, INSPECT AND CLEAN THE DATA

### Read the dataframe and store it in the variable "fifa"

```python
fifa=pd.read_csv("fifa_19.csv")
fifa
```

### Inspect the dataframe

```python
fifa.info()
```

```python
fifa.columns
```

### Drop unnecessary columns

```python
drop_list=[0,1,4,6,10,13,15,16,17,18,19,20]+[i for i in range(22,88)]
fifa.drop(fifa.columns[drop_list],axis=1,inplace=True)
fifa
```

### Data cleaning

```python
#Total null values in each column
fifa.isnull().sum()
```

```python
#Players with null values in club are called free agents
fifa['Club'].fillna('Free Agent',inplace=True)
```

```python
#Players with null values in release clause are replaces with €100K base price
fifa['Release Clause'].fillna('€100K',inplace=True)
```

```python
#Players with null values in preferred foot are replaced with both
fifa['Preferred Foot'].fillna('Both',inplace=True)
```

```python
#Players with null values in position are dropped from the data
fifa.dropna(inplace=True)
```

```python
#Total null values after cleaning
fifa.isnull().sum()
```

### Inspect the cleaned and usable dataframe

```python
fifa.info()
```

## TASK 2: EXPLORATORY DATA ANALYSIS

 ### Convert Value, Wage and Release Clause in million and adding new columns to the dataframe.

```python
def conversion(a):
    a=a.replace("€","")
    if "M" in a:
        a=a.replace("M","")
        a=float(a)
        a=a*1000000
    else:
        a=a.replace("K","")
        a=float(a)
        a=a*1000
    return a
        
fifa['Value(in numerical)']=fifa['Value'].apply(conversion)
fifa['Wage(in numerical)']=fifa['Wage'].apply(conversion)
fifa['Release Clause(in numerical)']=fifa['Release Clause'].apply(conversion)
fifa
```

### Average Age Of Players of Clubs

```python
y=fifa.groupby("Club").mean()['Age'].values
x=np.array(fifa.groupby("Club").count().index)
plt.figure(figsize=(18,7))
plt.ylabel("Average Age")
plt.xlabel("Teams")
plt.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
plt.scatter(x,y,color="purple",alpha=0.8)
```

#### From the graph we get an idea of team's average age of players and most teams has around the range 23-26 average age.
